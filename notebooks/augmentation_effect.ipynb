```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
delta_e_5_no_augment = np.load('../delta_e_run_2020_08_16_19_07_1597597658_False.npy')
delta_e_5_augment = np.load('../delta_e_run_2020_08_17_00_42_1597617725_True.npy')

delta_e_5_no_augment_test = np.load('../delta_e_run_2020_08_16_19_07_1597597658_False_test.npy')
delta_e_5_augment_test = np.load('../delta_e_run_2020_08_17_00_42_1597617725_True_test.npy')
```

```python
delta_e_16_augments = np.load('../delta_e_run_2020_08_17_00_42_1597617725_True.npy')
delta_e_16 = np.load('../delta_e_run_2020_08_17_14_13_1597666430_False.npy')

delta_e_16_augments_test = np.load('../delta_e_run_2020_08_18_03_03_1597712630_True_test.npy')
delta_e_16_test = np.load('../delta_e_run_2020_08_18_03_03_1597712630_True.npy')
```

```python
delta_e_not_no_augment = np.load('../delta_e_run_2020_08_18_20_31_1597775503_False.npy')
delta_e_not_augment = np.load('../delta_e_run_2020_08_20_04_18_1597889893_True.npy')

delta_e_not_no_augment_test = np.load('../delta_e_run_2020_08_18_20_31_1597775503_False_test.npy')
delta_e_not_augment_test = np.load('../delta_e_run_2020_08_20_04_18_1597889893_True_test.npy')
```

```python
fig, ax = plt.subplots(3,2, sharex='all', figsize=(8,8))

ax[0][0].hist(delta_e_5_augment.mean(axis=1), label='trained on multiple RGB/label', density=True)
ax[0][0].hist(delta_e_5_no_augment.mean(axis=1), label='trained on median RGB', density=True)

ax[0][1].hist(delta_e_5_augment_test.mean(axis=1), density=True)
ax[0][1].hist(delta_e_5_no_augment_test.mean(axis=1), density=True)

ax[1][0].hist(delta_e_16_augments.mean(axis=1), density=True)
ax[1][0].hist(delta_e_16.mean(axis=1), density=True)

ax[1][1].hist(delta_e_16_augments_test.mean(axis=1), density=True)
ax[1][1].hist(delta_e_16_test.mean(axis=1), density=True)


ax[2][0].hist(delta_e_not_augment.mean(axis=1), density=True)
ax[2][0].hist(delta_e_not_no_augment.mean(axis=1), density=True)

ax[2][1].hist(delta_e_not_augment_test.mean(axis=1), density=True)
ax[2][1].hist(delta_e_not_no_augment_test.mean(axis=1), density=True)


ax[0][0].legend()

for b in ax:
    for a in b:
        a.spines['left'].set_smart_bounds(True)
        a.spines['bottom'].set_smart_bounds(True)

    
ax[2][0].set_xlabel(r'$\Delta E^*_{ab}$')
ax[2][1].set_xlabel(r'$\Delta E^*_{ab}$')


ax[0][0].set_ylabel(r'$\sigma\left(\Delta E^*_{ab}\right) < 5$')
ax[1][0].set_ylabel(r'$\sigma\left(\Delta E^*_{ab}\right) < 16$')
ax[2][0].set_ylabel('no threshold')


ax[0][0].set_title('holdout test set from same distribution')
ax[0][1].set_title('holdout test set from full distribution')

fig.tight_layout()

fig.savefig('threshold_aug_effect.pdf', bbox_inches='tight')
```
