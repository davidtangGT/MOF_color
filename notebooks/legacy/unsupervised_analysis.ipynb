```python
import numpy as np 
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
X_train = np.load('../results/gbm/X_train_run_2020_06_08_12_59_1591613951.npy')
```

```python
y_train = np.load('../results/gbm/y_train_run_2020_06_08_12_59_1591613951.npy')
```

```python
import umap
reducer = umap.UMAP()
```

```python
X_transformed = reducer.fit_transform(X_train)
```

```python
plt.scatter(X_transformed[:,0], X_transformed[:,1], c=y_train)
```

## Liumapar projection

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
```

```python
X_transformed = pca.fit_transform(X_train)
```

```python
plt.scatter(X_transformed[:,0], X_transformed[:,1], c=y_train)
plt.xlim(0,100)
plt.ylim(-60,60)
```

## Non-linear method

```python
from sklearn.manifold import Isomap
```

```python
isomap = Isomap()
```

```python
X_isomap_transformed = isomap.fit_transform(X_train)
```

```python
plt.scatter(X_isomap_transformed[:,0], X_isomap_transformed[:,1], c=y_train)
```
