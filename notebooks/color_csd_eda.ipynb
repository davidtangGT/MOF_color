```python
import pickle
import pandas as pd
```

```python
with open('../../../proj73_cleancore/find_duplicates/compositions.pkl', 'rb') as fh:
    compositions = pickle.load(fh)
```

```python
compositions['ABAVIJ_clean'].as_dict()
```

```python
results = []

for k, v in compositions.items(): 
    res = v.as_dict()
    res['name'] = k
    res['refcode'] = k.split('_')[0]
    results.append(res)
```

```python
df_composition = pd.DataFrame(results)
```

```python
df_composition_filled = df_composition.fillna(0)
```

```python
df_composition_filled_in_color = df_composition_filled[df_composition_filled['refcode'].isin(list(df.refcode.values))]
```

```python
df_composition_filled_in_color['V'].unique()
```

```python
all_refcodes = list(df_train['refcode'].values) + list(df_test['refcode'].values)
```

```python
from collections import defaultdict
```

```python
composition_dict = defaultdict(list)
```

```python
for i, row in df_composition_filled_in_color.iterrows():
    d = dict(row)
    if row['refcode'] in all_refcodes:
        for k, v in d.items():
            if k not in ['name', 'refcode']: 
                if v > 0:
                    composition_dict[k].append(k)
```

```python
count_dict = {}

for k, v in composition_dict.items(): 
    count_dict[k] = len(v)
```

```python
sorted_dict = {key: value for key, value in sorted(count_dict.items(), key=lambda item: item[1], reverse=True)}
```

```python
print_list = ['{}: {}'.format(k, v) for k, v in sorted_dict.items()]
```
